```python
# Install required libraries
!pip install rasterio matplotlib numpy opencv-python pillow

# Import everything
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image  # <-- Fix: Add this import
import cv2

# Function to inspect the image
def inspect_image(file_path):
    try:
        with rasterio.open(file_path) as src:
            print("Image metadata:")
            print(f"Shape (height, width): {src.shape}")
            print(f"Number of bands: {src.count}")
            print(f"Data type: {src.dtypes[0]}")
            print(f"Min/Max values: {np.min(src.read(1)), np.max(src.read(1))}")

            # Plot the image
            plt.figure(figsize=(10, 10))
            plt.imshow(src.read(1), cmap='gray', vmin=np.percentile(src.read(1), 2), vmax=np.percentile(src.read(1), 98))
            plt.colorbar()
            plt.title("Raw Image (Auto-Stretched)")
            plt.show()
    except Exception as e:
        print(f"Error: {e}")

# Function to normalize to 8-bit
def normalize_to_8bit(file_path):
    with rasterio.open(file_path) as src:
        img = src.read(1)
        img_normalized = ((img - np.min(img)) / (np.max(img) - np.min(img)) * 255).astype(np.uint8)

        # Display
        plt.figure(figsize=(10, 10))
        plt.imshow(img_normalized, cmap='gray')
        plt.colorbar()
        plt.title("Normalized 8-bit Image")
        plt.show()

        # Save and download
        Image.fromarray(img_normalized).save("normalized.png")
        files.download("normalized.png")

# Function to enhance contrast (CLAHE)
def enhance_contrast(file_path):
    with rasterio.open(file_path) as src:
        img = src.read(1)
        img_8bit = ((img - np.min(img)) / (np.max(img) - np.min(img)) * 255).astype(np.uint8)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        img_enhanced = clahe.apply(img_8bit)

        # Display
        plt.figure(figsize=(10, 10))
        plt.imshow(img_enhanced, cmap='gray')
        plt.title("Contrast-Enhanced Image (CLAHE)")
        plt.colorbar()
        plt.show()

        # Save and download
        Image.fromarray(img_enhanced).save("enhanced.png")
        files.download("enhanced.png")

# Upload your file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Run all steps
inspect_image(file_name)
normalize_to_8bit(file_name)
enhance_contrast(file_name)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from google.colab import files

def process_image(file_path):
    with rasterio.open(file_path) as src:
        img = src.read(1)

        # Step 1: Clip negative values and scale
        img = np.clip(img, 0, None)
        vmin, vmax = np.percentile(img, [2, 98])
        img_normalized = ((img - vmin) / (vmax - vmin) * 255).astype(np.uint8)

        # Step 2: CLAHE with softer settings
        clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(16, 16))
        img_enhanced = clahe.apply(img_normalized)

        # Plot results
        fig, axes = plt.subplots(1, 3, figsize=(18, 6))
        axes[0].imshow(img, cmap='gray', vmin=vmin, vmax=vmax)
        axes[0].set_title("Raw (Clipped)")
        axes[1].imshow(img_normalized, cmap='gray')
        axes[1].set_title("Normalized (0–255)")
        axes[2].imshow(img_enhanced, cmap='gray')
        axes[2].set_title("CLAHE Enhanced")
        plt.show()

        # Save enhanced image
        Image.fromarray(img_enhanced).save("enhanced.png")
        files.download("enhanced.png")

# Upload and process
uploaded = files.upload()
process_image(list(uploaded.keys())[0])
```
